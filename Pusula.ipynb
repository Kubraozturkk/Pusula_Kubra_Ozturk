```python
# Exploratory Data Analysis (EDA) ile elde ettiğim ön değerlendirme sonucunda:
# 1. İl sütununda eksik değerlerin en sık görülen değerle doldurulması planlanmıştır.
# 2. Uyruk herkes için aynı olduğundan dolayı uyruk sütunu silinecektir.
# 3. Cinsiyet sütununda çok fazla eksik veri olduğundan dolayı cinsiyet sütunu silinecektir. Daha sonra eldeki eksiksiz veriler ve diğer sütunlar kullanılarak bir cinsiyet tahmin modeli geliştirilebilir.
# 4. İlaç başlangıç ve bitiş tarihlerinden ilaç kullanım süresi olarak yeni bir sütun oluşturulacak.
# 5. Kilo ve boy sütunlarında eksik değerler ortalama alınarak doldurulacaktır.
# 6. Kilo ve boy verilerinden vücut kitle indeksi hesaplanacak, vki sütunu oluşturulacak.
# 7. Kronik hastalıklar sütunundaki eksik değerler en çok tekrar değerlerle doldurulacaktır.
# 8. Doğum tarihi sütunundan yaş hesaplanacak ve yaş adında yeni bir sütun oluşturulacak.
# 9. İlaca başlama ve yan etki bildirme süresinden yola çıkılarak yan etki oluşum süresi hesaplanarak yan etki oluşum süresi adında yeni bir sütun oluşturulacaktır.
# 10. Kan grubu sütunundaki eksik değerlerin en sık görülen değerle doldurulması planlanmıştır.
# 11. Alerji sütunundaki eksik değerlerin en sık görülen değerle doldurulması planlanmıştır.
# 12. Kullanıcı id sütunu silinecektir.
```

# **Preprocessing**

```python
import numpy as np
```

```python
# İl sütununda eksik değerlerin en sık görülen değerle doldurulması
most_frequent_value = df['Il'].mode()[0]
```

```python
# İl sütununda eksik değerlerin en sık görülen değerle doldurulması
df['Il'].fillna(most_frequent_value, inplace=True)
```

```python
# Uyruk sütununun silinmesi
df.drop(columns=['Uyruk'], inplace=True)
```

```python
# Kullanıcı id sütununun silinmesi
df.drop(columns=['Kullanici_id'], inplace=True)
```

```python
# Cinsiyet sütununun silinmesi
df.drop(columns=['Cinsiyet'], inplace=True)
```

```python
# Boy ve kilo değerlerinin ortalamalarını alınması
df['Kilo'].fillna(df['Kilo'].mean(), inplace=True)  # Ortalama ile doldur
df['Boy'].fillna(df['Boy'].mean(), inplace=True)    # Ortalama ile doldur
```

```python
# İlaç kullanım süresi hesaplama
df['ilaç_kullanım_süresi(gün)'] = (pd.to_datetime(df['Ilac_Bitis_Tarihi']) - pd.to_datetime(df['Ilac_Baslangic_Tarihi'])).dt.days
```

```python
# Yan etki oluşum süresi hesaplama
df['Ilac_Baslangic_Tarihi'] = pd.to_datetime(df['Ilac_Baslangic_Tarihi'])
df['Yan_Etki_Bildirim_Tarihi'] = pd.to_datetime(df['Yan_Etki_Bildirim_Tarihi'])
df['Yan_Etki_Olusum_Suresi(Gün)'] = (df['Yan_Etki_Bildirim_Tarihi'] - df['Ilac_Baslangic_Tarihi']).dt.days
```

```python
# Vücut Kitle İndeksi (vki) hesaplama
df['vki'] = df['Kilo'] / ((df['Boy'] / 100) ** 2)
```

```python
df = df[df['Kan Grubu'] != 'Bilinmiyor']
```

```python
most_frequent_kan_grubu = df['Kan Grubu'].mode()[0]
```

```python
df['Kan Grubu'].fillna(most_frequent_kan_grubu, inplace=True)
```

```python
most_frequent_alergi = df['Alerjilerim'].mode()[0]
df['Alerjilerim'].fillna(most_frequent_alergi, inplace=True)
```

```python
# Her sütundaki eksik değerlerin sayısını yazdırma
print(df.isnull().sum())
```

```python
from datetime import datetime

# 2024 yılına göre yaş hesaplama
current_year = 2024
df['Yaş'] = current_year - df['Dogum_Tarihi'].dt.year
```
